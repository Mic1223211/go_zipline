```python
%%zipline --start 2014-1-1 --end 2018-1-1
from zipline.api import order_target,record,symbol,order_target_percent
from zipline.algorithm import TradingAlgorithm
from zipline.finance import commission
from sklearn.ensemble import RandomForestClassifier
from collections import deque
import matplotlib.pyplot as plt
import pyfolio as pf
import sys
import numpy as np
from datetime import datetime
import pytz

def initialize(context):
    context.security = symbol('AAPL')
    context.window_length = 10 
    
    context.classifier = RandomForestClassifier() 
    
    context.recent_prices = deque(maxlen=context.window_length+2) 
    context.X = deque(maxlen=500) 
    context.Y = deque(maxlen=500) 
    
    context.prediction = 0 # Stores most recent prediction
    
def handle_data(context,data):
    context.recent_prices.append(data.current(context.security, 'price')) 
    if len(context.recent_prices) == context.window_length+2: 
        
        changes = np.diff(context.recent_prices) > 0
        
        newChs = [];
        for index in range(len(changes)):
            if changes[index]: # True
                newChs.append(1)
            else:
                newChs.append(0)
        context.X.append(newChs[:-1]) 
        context.Y.append(newChs[-1]) 
        
        if len(context.Y) >= 100: 
            
            context.classifier.fit(context.X, context.Y)
            test = changes[1:].reshape(1, -1)
            
            context.prediction = context.classifier.predict(test) # Predict
            if(context.prediction > 0):
                order_target(context.security,100)
            else:
                order_target(context.security,-10)
                
            
    record(prediction=int(context.prediction))
    
def analyze(context,perf):
    fig = plt.figure()
    ax1 = fig.add_subplot(211)
    perf.portfolio_value.plot(ax=ax1)
    ax1.set_ylabel('portfolio value in $')
    
```
